**In this notebook, we compare the effects of pH and LA binding at 10ns through bar graphs.**

We do this for key segments which have shown to show the greatest response to pH and LA-binding including:

(1) The N-Terminal Domain, NTD (S71-R78, H146-E156,L249-G257); 
(2) The Receptor Binding Motif, RBM; 
(3) The Furin Cleavage Site, FS; 
(4) The V622-L629 Segment;
(5) The D803-S813 Segment; 
(6) The Fusion-Peptide Proximal Region, FPPR.

The data used for plotting has determined using the peak deviation analysis script. To investigate this, refer to 'peak_deviation_calculator'.

**(1) The NTD Domain**

**(1.1) The NTD Domain: S71-R78 Segment**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data for the first graph
data1 = {
    'Experimental Condition': ['pH 5', 'pH 5', 'pH 10', 'pH 10'],
    'Cα Deviation': [0.486, 0.437, 0.498, 0.412],
    'Condition': ['Apo', 'LA-Bound', 'Apo', 'LA-Bound']
}

df1 = pd.DataFrame(data1)

# Data for the second graph
data2 = {
    'Spike Condition': ['Apo', 'Apo', 'LA-Bound', 'LA-Bound'],
    'Cα Deviation': [0.486, 0.498, 0.437, 0.412],
    'pH Condition': ['pH 5', 'pH 10', 'pH 5', 'pH 10']
}

df2 = pd.DataFrame(data2)

# Create a figure with two wider subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6))  # Adjusted figsize to increase space

# Common font size for titles and labels
common_fontsize = 14
legend_fontsize = 10  

# Plot the first graph
sns.barplot(ax=ax1, x='Experimental Condition', y='Cα Deviation', hue='Condition', data=df1, palette=['#4682B4', '#FA8072'])  
ax1.set_xlabel('pH Condition', fontweight='bold', fontsize=common_fontsize)
ax1.set_ylabel('Cα Deviation (nm)', fontweight='bold', fontsize=common_fontsize)
ax1.set_title('S71-R78: Cα Deviation Across pH for Apo and LA-Bound States', fontweight='bold', fontsize=common_fontsize, pad=15)  
ax1.set_ylim(0, 0.7)  # Set y-limit
ax1.legend(loc='upper right', bbox_to_anchor=(1.0, 1.0), title="Spike Condition", fontsize=legend_fontsize)

# Annotate the bars with the deviation values
for p in ax1.patches:
    ax1.annotate(f'{p.get_height():.3f}', 
                 (p.get_x() + p.get_width() / 2., p.get_height()), 
                 ha = 'center', va = 'center', 
                 xytext = (0, 9), 
                 textcoords = 'offset points')

# Plot the second graph
sns.barplot(ax=ax2, x='Spike Condition', y='Cα Deviation', hue='pH Condition', data=df2, palette=['#3CB371', '#B0C4DE'])  
ax2.set_xlabel('Spike Condition', fontweight='bold', fontsize=common_fontsize)
ax2.set_ylabel('Cα Deviation (nm)', fontweight='bold', fontsize=common_fontsize)
ax2.set_title('S71-R78: Cα Deviation Across Spike Conditions at Different pH Levels', fontweight='bold', fontsize=common_fontsize, pad=15)  
ax2.set_ylim(0, 0.7)  # Set y-limit
ax2.legend(loc='upper right', bbox_to_anchor=(1.0, 1.0), title="pH Condition", fontsize=legend_fontsize)

# Annotate the bars with the deviation values
for p in ax2.patches:
    ax2.annotate(f'{p.get_height():.3f}', 
                 (p.get_x() + p.get_width() / 2., p.get_height()), 
                 ha = 'center', va = 'center', 
                 xytext = (0, 9), 
                 textcoords = 'offset points')

# Adjust layout with increased horizontal space between plots
plt.subplots_adjust(wspace=0.6)  # Increase the width space between subplots
plt.tight_layout(rect=[0, 0, 0.85, 1])  # Adjust plot area to make space for the legend

# Save the plot
plt.savefig('S71-R78 comparison bars.png')
plt.show()


```
