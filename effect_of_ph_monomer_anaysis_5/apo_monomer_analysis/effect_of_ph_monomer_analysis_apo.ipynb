**In this notebook we analyse the response of each monomer individually to pH for the Apo condition through bar graphs.**

We do this for key segments which have shown to show the greatest response to pH whilst not bound to LA (Apo) including:

(1) The N-Terminal Domain, NTD (S71-R78, H146-E156,L249-G257); 
(2) The Receptor Binding Motif, RBM; 
(3) The Furin Cleavage Site, FS; 
(4) The V622-L629 Segment;
(5) The D803-S813 Segment; 
(6) The Fusion-Peptide Proximal Region, FPPR.

The data used for plotting has determined using the peak deviation analysis script. To investigate this, refer to 'peak_deviation_calculator'.

**(1) The NTD Domain**

**(1.1) S71-R78**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Monomer': ['Monomer A', 'Monomer A', 'Monomer B', 'Monomer B', 'Monomer C', 'Monomer C'],
    'Cα Deviation': [0.505, 0.533, 0.442, 0.461, 0.512, 0.501], 
    'pH Condition': ['pH 5', 'pH 10', 'pH 5', 'pH 10', 'pH 5', 'pH 10']
}

# Convert the data into a DataFrame
df = pd.DataFrame(data)

# Set plot style
sns.set(style="whitegrid")

# Common font size for the title, x-axis, and y-axis
common_fontsize = 14

# Create the bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='Monomer', y='Cα Deviation', hue='pH Condition', data=df, palette=['#3CB371', '#B0C4DE'])  # Medium Sea Green and Light Steel Blue

# Customize the plot with common font size
plt.xlabel('Monomer', fontweight='bold', fontsize=common_fontsize)
plt.ylabel('Cα Deviation (nm)', fontweight='bold', fontsize=common_fontsize)
plt.title('S71-R78: Cα Deviation Across Monomers at Apo pH 5 and pH 10', fontweight='bold', fontsize=common_fontsize, pad=15)
plt.ylim(0, 0.7)  
plt.legend(title='pH Condition', loc='upper right', fontsize=10)

# Annotate the bars with deviation values
for p in plt.gca().patches:
    plt.annotate(f'{p.get_height():.3f}', 
                 (p.get_x() + p.get_width() / 2., p.get_height()), 
                 ha='center', va='center', 
                 xytext=(0, 9), 
                 textcoords='offset points')

# Adjust layout
plt.tight_layout()

# Save the plot
plt.savefig('S71-R78_Apo_Monomer_Analysis.png', dpi=300)  # Save the plot as a PNG file with high resolution

# Show the plot
plt.show()
```
